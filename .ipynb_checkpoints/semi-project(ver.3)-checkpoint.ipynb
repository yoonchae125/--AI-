```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from konlpy.tag import Twitter
from collections import Counter

from bs4 import BeautifulSoup
import time
from datetime import datetime

import requests
import pandas as pd
import nltk
import pickle
```

```python
comments_list = [] # 댓글 리스트 먼저 만들기


driver = webdriver.Chrome(executable_path='(driver) chromedriver.exe') # 가상 브라우저 열기
urls_list = ['https://news.v.daum.net/v/20191216173753256', 'https://news.v.daum.net/v/20191216152615407', 'https://news.v.daum.net/v/20191216133908705', 'https://news.v.daum.net/v/20191216163405068', 'https://news.v.daum.net/v/20191216155336614', 'https://news.v.daum.net/v/20191216171142329', 'https://news.v.daum.net/v/20191216175415829', 'https://news.v.daum.net/v/20191216145822208', 'https://news.v.daum.net/v/20191216145426980', 'https://news.v.daum.net/v/20191216162019633']

for i in range(len(urls_list)): # i가 뭔지 잘 모르겠다;;

    driver.get(urls_list[i]) 

    finish_flag = 1 # 댓글 더보기 버튼 클릭 횟수 제한하기
    while finish_flag == 1:
        try:
            driver.find_element_by_class_name('link_fold').click()
            time.sleep(2) # ip차단 방지
        except:
            finish_flag = 0 # false일 경우 멈추고 아래로 이동.
        
    source = driver.page_source
    source = BeautifulSoup(source, 'html.parser')  # try 라인에 맞춰 들여쓰기 했을 때 같은 댓글이 반복되어 저장되는 오류 발생

    for tag in source.find_all('p', {'class':'desc_txt'}):
        comments_list.append(tag.get_text())
    
    print('Finished crawling on : ', urls_list[i])
    
driver.close()
driver.quit()
```

```python
with open('raw_comments.txt', 'rb') as f:
    raw_comments = pickle.load(f)
```

```python
len(raw_comments)
```

```python
raw_comments_1 = raw_comments[0:500]
raw_comments_2 = raw_comments[500:]
```

```python
pos_result = []

twitter = Twitter()

for idx, comment in enumerate(raw_comments_1):
    if idx % 100 == 0:
        print('{}번째 댓글 태깅 완료'.format(idx))
    else:
        pos_result.extend(twitter.pos(comment, norm=True, stem=True))
        
with open('pos_result_1.txt', 'wb') as f:
    pickle.dump(pos_result, f)
```

```python
pos_result = []

for idx, comment in enumerate(raw_comments_2):
    if idx % 100 == 0: # 왜 이렇게 쓰셨지.....?
        print('{}번째 댓글 태깅 완료'.format(idx))
    else:
        pos_result.extend(twitter.pos(comment, norm=True, stem=True))
        

with open('pos_result_2.txt', 'wb') as f:
    pickle.dump(pos_result, f)
```

```python
with open('pos_result_1.txt', 'rb') as f:
    pos_result_1 = pickle.load(f)
    
with open('pos_result_2.txt', 'rb') as f:
    pos_result_2 = pickle.load(f)
```

```python
pos_result = pos_result_1 + pos_result_2
pos_result
```

```python
len(pos_result)
```

```python
del_list = ['하다', '있다', '되다', '이다', '돼다', '않다', '그렇다', '아니다', '이렇다', '그렇다', '어떻다'] 

comments_cleaned = []

('인기', 'Noun')    
for comment in pos_result: #  raw_pos_tagged= ('서울', 'Noun') 형식
    
    if comment[1] not in ["Josa", "Eomi", "Punctuation", "Foreign"]: # Foreign == ”, “ 와 같이 제외되어야할 항목들
        # not은 word의 앞이나 뒤나 상관없음. 
        # word[0]이 아니라 word[1]이 들어가는 이유는 주어진 튜플 형식(단어, 품사)에서 품사의 위치가 두번째(1)이기 때문!! 
        if (len(comment[0])> 1) & (comment[0] not in del_list): # 한 글자로 이뤄진 단어들을 제외 & 원치 않는 단어들을 제외
            # &는 and로 변환 가능!
            # 질문! !=1이 아니라 >1 같은 식으로는 안 되나? -> 됨. 형식은 !를 빼고 >1 혹은 2와 같거나 크다(부등호) 삽입.
            # '!=' 이 기호는 '~가 아니다'라는 뜻이라고.
            comments_cleaned.append(comment[0])
        
comments_cleaned
```

```python
from collections import Counter

result = Counter(comments_cleaned) # Counter의 기능은 숫자세기라고 생각하면 된다! #타입명이기도 함
comments_dic = dict(result)
comments_dic
```

```python
comments_dic.items()
```

```python
sorted_comments_dic = sorted(comments_dic.items(), key=lambda x:x[1], reverse=True)
sorted_comments_dic
```

```python
for comment, count in sorted_comments_dic[:100]:
    print("{0}({1})".format(comment, count), end=" ")
```
