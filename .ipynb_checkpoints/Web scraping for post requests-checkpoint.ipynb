```python
data = {'stdrYyCd': '2019', 'stdrQuCd': '2', 'svcIndutyCdL': 'CS000000', 'svcIndutyCdM': 'all'}
```

```python
result = requests.post('https://golmok.seoul.go.kr/region/selectRentalPrice.json', data=data).text
```

```python
result = json.loads(result)
result
```
