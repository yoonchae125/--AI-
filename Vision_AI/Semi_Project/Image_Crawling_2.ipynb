```python
import cv2
from icrawler.builtin import GoogleImageCrawler
import os
download_dir="img"
if not os.path.exists(download_dir):
    os.makedirs(download_dir)
#     print("make directory")
import re
import glob
import numpy as np
from pathlib import Path
```

```python
print("원하는 그림의 키워드를 넣어주세요: ")
# kw=input()
kw = '쯔위'
```

```python
needless = ['\'', '[', ']', ',']
str_kw = str(kw)
for i in needless:
    str_kw = str_kw.replace(i, '')
download_dir += '/' + str_kw
print(download_dir)
```

```python
Path(download_dir).mkdir(parents=True, exist_ok=True)
google_crawler=GoogleImageCrawler(
    feeder_threads=1,
    parser_threads=1,
    downloader_threads=4,
    storage={'root_dir':download_dir})
filters=dict(size='small', #이미지 크기
            color='color',#이미지의 색깔을 필터링
            license='commercial,modify',
            )#검색하고자 하는 날짜의 범위

# type='animated',
# date=((2017,1,1),(2017,12,31))
# {'date':((2016,1,1),(2016,6,30))}
google_crawler.crawl(keyword=kw,
                     filters=None,
                     max_num=1500,
                     file_idx_offset=0)
```

```python
def my_imread(filename, flags=cv2.IMREAD_COLOR, dtype=np.uint8):
    try:
        n = np.fromfile(filename, dtype)
        img = cv2.imdecode(n, flags)
        return img
    except Exception as e:
        print(e)
        return None
```

```python
def my_imwrite(filename, img, params=None):
    try:
        ext = os.path.splitext(filename)[1]
        result, n = cv2.imencode(ext, img, params)
        if result:
            with open(filename, mode='w+b') as f:
                n.tofile(f)
            return True
        else:
            return False
    except Exception as e:
        print(e)
        return False
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#폴더 내의 이미지 파일명 리스트 반환
file_list = os.listdir(download_dir)
```

```python
#각 이미지 얼굴만 추출저장
for idx, f_name in enumerate(file_list):
    try:
        print('processing: ' + str(idx))
        #file_type = re.findall('[.]\S+', f_name)[0]
        f_name1 = f_name
        f_name = download_dir + '/' + f_name
        img = my_imread(f_name)
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 7)
        if type(faces) == tuple:
            continue
        tmp_px=20
        h_tmp_px=tmp_px//2
        for (x,y,w,h) in faces:
            img = cv2.rectangle(img,(x-h_tmp_px,y-h_tmp_px),(x+w+h_tmp_px,y+h+h_tmp_px),(255,0,0),2)
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]
        roi_color=cv2.resize(roi_color,(100,100),interpolation=cv2.INTER_LINEAR)
        save_dir = download_dir + '_face'
        w_name = save_dir + '/' + f_name1
        if not os.path.exists(save_dir):
            os.makedirs(save_dir)
        my_imwrite(w_name, roi_color)
    except:
        pass
print("done")
```
